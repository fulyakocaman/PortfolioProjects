```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
```

## Experiments 1 and 2: 
Download the bank-additional.csv for training and bank-additional-full.csv for testing, and then examine each dataset.

```python
dfTrainingSet = pd.read_csv('bank-additional.csv', sep=';')
dfTrainingSet
```

```python
dfTestSet = pd.read_csv('bank-additional-full.csv', sep=';')
dfTestSet
```

## Experiment 3: 
The training and test DataFrames will need some significant preprocessing before they can be used:

a. Several of the features are categorical variables and will need to be turned into numbers before they can be used by ML algorithms. The simplest way to accomplish this is to use dummy coding using the Pandas get_dummies() method. Note: Some algorithms (e.g. logistic regression) have problems with collinear features. If you use one-hot encoding, one dummy variable will be a linear combination of the other dummy variables, so be sure to pass drop_first=True.

b. Two features should be removed from the dataset:
i. Per bank-additional-names.txt, the duration “should be discarded if the intention is to have a realistic predictive model.”
ii.The feature y (or y_yes after dummy coding) is the target.

c. Some algorithms (e.g. KNN and SVM) require non-categorical features to be standardized. 


### The Result of Experiment 3 (a): 

```python
dfNewTrainingSet = pd.get_dummies(dfTrainingSet, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome', 'y'], drop_first=True)
dfNewTrainingSet
```

```python
dfNewTestSet = pd.get_dummies(dfTestSet, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome', 'y'], drop_first=True)
dfNewTestSet
```

### The Result of Experiment 3 (b): 

```python
dfNewTrainingSet.drop('duration', inplace=True, axis=1)
dfNewTrainingSet
```

```python
# Storing the feature y_yes as the target before removing it from the dataframe
targetTraining = dfNewTrainingSet['y_yes']
```

```python
dfNewTrainingSet.drop('y_yes', inplace=True, axis=1)
dfNewTrainingSet
```

```python
dfNewTestSet.drop('duration', inplace=True, axis=1)
dfNewTestSet
```

```python
# Storing the feature y_yes as the target before removing it from the dataframe
targetTest = dfNewTestSet['y_yes']
```

```python
dfNewTestSet.drop('y_yes', inplace=True, axis=1)
```

```python
dfNewTestSet
```

### The Result of Experiment 3 (c): 

```python
scaler= StandardScaler()
# standardization of feture variables which are now all non-categorical features
scaled_TrainingSet = scaler.fit_transform(dfNewTrainingSet)
scaled_TestSet = scaler.fit_transform(dfNewTestSet)
```

## Experiment 4: 
Fit Naive Bayes, KNN, and SVM classifiers to the training set, then score each classifier on the test set. Which classifier has the highest accuracy?

```python
gnbTraining = GaussianNB().fit(scaled_TrainingSet, targetTraining)
gnbTraining.score(scaled_TestSet,targetTest)
```

```python
neighTraining = KNeighborsClassifier(n_neighbors=3).fit(scaled_TrainingSet, targetTraining)
neighTraining.score(scaled_TestSet,targetTest)
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svmTraining = clf.fit(scaled_TrainingSet, targetTraining)

svmTraining.score(scaled_TestSet,targetTest)
```

## The result of Experiment 4: 

According to the scores I got above, Naive Bayes, KNN, and SVM classifiers scored 88.73%, 88.67%, and 89.82% respectively. Therefore, SVM classifier has the highest accuracy among the other classifiers.

## Experiment 8: 
One of the easiest ways to deal with an unbalanced dataset is random oversampling. This can be done with an imblearn.over_sampling.RandomOverSampler object. Use fit_resample() to generate a balanced training set. To make sure that your results are reproducible, pass random_state=(2021-4-22).

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=2021-4-22)

resampledTraining, resampledTargetTraining = ros.fit_resample(scaled_TrainingSet, targetTraining)
```

## Experiment 9: 
Repeat experiments (4) and (7) on the balanced training set of experiment (8). Which classifier performs the best, and how much better is its performance?

```python
gnbResampled = GaussianNB().fit(resampledTraining, resampledTargetTraining)
gnbResampled.score(scaled_TestSet,targetTest)
```

```python
#prediction on the fitted model from Naive Bayes Classifier
y_pred1 = gnbResampled.predict(scaled_TestSet)

confusion_matrix(targetTest, y_pred1)
```

```python
roc_auc_score(targetTest, y_pred1)
```

```python
neighResampled = KNeighborsClassifier(n_neighbors=3).fit(resampledTraining, resampledTargetTraining)
neighResampled.score(scaled_TestSet,targetTest)
```

```python
#prediction on the fitted model from KNN Classifier
y_pred2 = neighResampled.predict(scaled_TestSet)

confusion_matrix(targetTest, y_pred2)
```

```python
roc_auc_score(targetTest, y_pred2)
```
